```python
import matplotlib.pyplot as plt
n = int(input("Enter the number of points: "))
x = []
y = []
for i in range(0, n):
    xco = int(input("Enter xco: "))
    yco = int(input("Enter yco: "))
    x.append(xco)
    y.append(yco)
print(x)
print(y)
plt.plot(x,y,'go--')
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = (np.arange(0,360,1)*np.pi)/180
y = np.sin(x)
print(y)
plt.grid()
plt.plot(x,y)
plt.show()
```
